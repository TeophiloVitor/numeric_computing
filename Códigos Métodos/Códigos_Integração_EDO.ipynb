```python
#Interpolação de Lagrange
import numpy as np
def lagrange(x, y, xp):
    soma = 0
    for i in range(len(y)):
        produto = 1
        print(y[i], end="")
        for j in range(len(x)):
            if j != i:
                produto = produto*(xp-x[j])/(x[i] - x[j])
                print('*( xp -', x[j], ')/(', x[i], '-', x[j], ')', end="")
        print(' + ', end="")
        soma = soma + produto*y[i]
    print()
    return soma
texto = input('digite x: ')
a = texto.split(',')
texto = input('digite y: ')
b = texto.split(',')
xp = float(input('digite xp: '))
x = np.array([float(c) for c in a])
y = np.array([float(c) for c in b])
print('f(xp): ', "%.8f" % (lagrange(x, y, xp)))
```

```python
#Interpolação de Newton
import numpy as np
def newton(x, y, xp):
    soma = m[0, 0]
    print(m[0, 0], ' + ', end="")
    for i in range(1, len(x)):
        produto = 1
        print(m[i, 0], end="")
        for j in range(i):
            produto = produto*(xp-x[j])
            print('*( xp -', x[j], ')', end="")
        print(' + ', end="")
        soma = soma + m[i, 0]*produto
    print()
    return soma
texto = input('digite x: ')
a = texto.split(',')
texto = input('digite y: ')
b = texto.split(',')
xp = float(input('digite xp: '))
x = np.array([float(c) for c in a])
y = np.array([float(c) for c in b])
m = np.zeros([len(x), len(x)])
m[0, :] = y
for i in range(1, len(x)):
  for j in range(0, len(x)-i):
    m[i, j] = (m[i-1, j+1]-m[i-1, j])/(x[i+j]-x[j])
print('Tabela de diferenças divididas: \n', m.T)
print('f(xp): ', "%.8f" % (newton(x, y, xp)))
```

```python
#Integral trapézio vetores
import numpy as np
texto = input('digite x: ')
x = texto.split(',')
texto = input('digite y: ')
y = texto.split(',')
x = np.array([float(c) for c in x], dtype = float)
y = np.array([float(c) for c in y], dtype = float)
soma = 0
for i in range(1, len(x)):
  h = x[i] - x[i - 1]
  I = (h/2)*(y[i] + y[i - 1])
  print('(', h, '/2 )*(', y[i], '+', y[i - 1], ') +', end="")
  soma = soma + I
print()
print('I: ', "%.8f" % soma)
```

```python
#Integral trapézio composto
import numpy as np
def f(x):
  return x**2
a = float(input('digite a: '))
b = float(input('digite b: '))
n = int(input('digite n: '))
h = (b - a)/n
soma = 0
print('(', h, '/2 )*(', end="")
for i in range(n + 1):
  if i == 0 or i == n:
    soma = soma + f(a)
    print(f(a), '+', end="")
  else:
    soma = soma + 2*f(a)
    print('2*', f(a), '+', end="")
  a = a + h
print(')')
I = (h/2)*soma
print('I: ', "%.8f" % I)
```

```python
#EDO primeira ordem com Euler
import numpy as np
def f(x, y):
  return y
t0 = float(input('digite t0: '))
y0 = float(input('digite y0: '))
h = float(input('digite h: '))
n = int(input('digite n: '))
y = y0
x = t0
print('x', 0,':', x) 
print('y', 0, ':', y)
for i in range(n):
  aux = y
  y = y + h*f(x, y)
  x = x + h
  print('x', i + 1,':', x) 
  print('y', i + 1, ':', aux, '+', h, '*', aux)
  print('y', i + 1, ':', y)
print('y: ', "%.8f" % y)
```

```python
#EDO primeira ordem com Euler modificado (Heun)
import numpy as np
def f(x, y):
  return y
t0 = float(input('digite t0: '))
y0 = float(input('digite y0: '))
h = float(input('digite h: '))
n = int(input('digite n: '))
y = y0
x = t0
print('x', 0,':', x) 
print('y', 0, ':', y)
for i in range(n):
  aux = y
  k1 = f(x, y)
  print('k1:', k1)
  k2 = f(x + h, y + h*k1)
  print('k2:', k2)
  y = y + (h/2)*(k1 + k2)
  x = x + h
  print('x', i + 1,':', x) 
  print('y', i + 1, ':', aux, '+ (', h, '/2 )*(', k1, '+', k2, ')')
  print('y', i + 1, ':', y)
print('y: ', "%.8f" % y)
```
