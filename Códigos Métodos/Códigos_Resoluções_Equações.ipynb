```python
#Eliminação de Gauss
import numpy as np
def gauss(ma, n):
  for i in range(n):
    pivo = ma[i, i]
    for j in range(i + 1, n):
      mzinho = ma[j, i]/pivo
      print('FATOR', j, ',', i, 'É', mzinho)
      ma[j, :] = ma[j, :] - mzinho*ma[i, :]
    print('COLUNA ESCALONADA', i, '\n', ma)
  return ma
y = np.array([15, 26, 10], dtype = float)
m = np.array([[5, 2, 2], [2, 3, 6], [1, 3, 1]], dtype = float)
ma = np.concatenate((m,np.transpose([y])), axis = 1) 
print('MATRIZ AUMENTADA\n', ma)
gauss(ma, len(y))
print('MATRIZ ESCALONADA\n', ma)
```

```python
#Resolução Retroativa
import numpy as np
np.set_printoptions(precision = 8)
def resret(m, x):
  y = np.empty([len(x)], dtype = float)
  for i in range(len(x) - 1, -1, -1):
    soma = 0
    for j in range(i + 1, len(x)):
      soma = soma + m[i, j]*y[j]
    y[i] = (x[i] - soma)/m[i, i]
  return y
m = np.array([[-9, 8, 10], [0, 26, 26], [0, 0, 3.33333334]], dtype = float)
x = np.array([-22, -26, -65.33333334], dtype = float)
print('VETOR SOLUÇÃO', resret(m, x))
```

```python
#CONVERGENCIA JACOBI
def linhas(m):
  for i in range(3):
    soma = 0
    for j in range(3):
      if i != j:
        soma = soma + m[i, j]
    if soma >= m[i, i]:
      return 0
  return 1
def colunas(m):
  ma = np.empty([3, 3])
  for i in range(3):
    for j in range(3):
      ma[i, j] = m[i, j]/m[i, i]
  for j in range(3):
    soma = 0
    for i in range(3):
      if i!= j:
        soma = soma + ma[i, j]
    if soma >= ma[j, j]:
      return 0
  return 1
C = np.array([[10, 2, 1], [1, 5, 1], [2, 3, 10]])
if colunas(C) == 1: 
  print("Converge para colunas")
if linhas(C) == 1:
  print("Converge para linhas")
else:
  print('Não converge')
```

```python
#CONVERGENCIA GAUSS-SEIDEL
def linhas(m):
  for i in range(3):
    soma = 0
    for j in range(3):
      if i != j:
        soma = soma + m[i, j]
    if soma >= m[i, i]:
      return 0
  return 1
def sassenfeld(m):
  ma = np.empty([3, 3])
  for i in range(3):
    for j in range(3):
      ma[i, j] = m[i, j]/m[i, i]
  somav = np.array([1]*3, dtype = float)
  for i in range(3):
    soma = 0
    for j in range(3):
      if i != j:
        soma = soma + ma[i, j]*somav[j]
    somav[i] = soma
    if soma >= m[i, i]:
      return 0
  return 1
C = np.array([[10, 2, 1], [1, 5, 1], [2, 3, 10]])
if sassenfeld(C) == 1: 
  print("Converge para sassenfeld")
if linhas(C) == 1:
  print("Converge para linhas")
else:
  print('Não converge')
```

```python
#JACOBI SISTEMAS
import numpy as np
def jacobi(m, y, x0, erro, itrmax):
  E = 1
  x = x0
  cont = 0
  while E > 10**(-erro) and cont < itrmax:
    xa = np.array(x)
    for i in range(len(y)):
      soma = 0
      for j in range(len(y)):
        if j != i:
          soma = soma + m[i, j]*xa[j] #PARA GAUSS-SEIDEL APAGUE O 'a'
      x[i] = (y[i] - soma)/m[i, i]
    print('VETOR', x)
    E = max(abs(x - xa))/max(abs(x))
    print('ERRO ', cont, E)
    cont = cont + 1
  return x, cont
erro = int(input('DIGITE A PRECISÃO '))
itrmax = int(input('DIGITE A ITRMAX '))
y = np.array([3, -8, 9], dtype = float)
m = np.array([[8, 2, -3], [-2, 7, 1], [-1, 3, 6]], dtype = float)
x0 = np.array([0]*len(y), dtype= float)
x, cont = jacobi(m, y, x0, erro, itrmax)
print('VETOR SOLUÇÃO ', x)
print('CONTADOR ', cont)
```

```python
#MétodoTrapézio
import math as ma
def motodot(f, x0, xM, h):
  inte = 0
  x = x0
  inte += eval(f)
  x += h
  while x < xM:
    inte += 2*eval(f)
    x += h
  x = xM
  inte += eval(f)
  inte *= (h/2)
  return inte

f = input()
x0 = float(input())
xM = float(input())
h = float(input())

motodot(f, x0, xM, h)
```

```python
#MétodoEuler
import numpy as np
import math as ma

xi = 2
xf = 4
yi = 1
h = 0.5

def edo(x,y):
  return ma.pow(x,2) - 3*y

def euler(xi,xf,yi,h):
  xn = xi
  yn = yi
  while xn < xf:
    yn1 = yn + h*edo(xn,yn)
    yn = yn1
    xn += h
  print("y(%.4f) = % .4f" % (xf, yn))

euler(xi,xf,yi,h)
```

```python
#MétodoEulerAperfe
import numpy as np
import math as ma

xi = 2
xf = 4
yi = 1
h = 0.5

def edo(x,y):
  return ma.pow(x,2) - 3*y

def euleraper(xi,xf,yi,h):
  xn = xi
  yn = yi
  xn1 = xn + h
  while xn < xf:
    k1 = edo(xn,yn)
    k2 = edo(xn1,yn + h*k1)
    yn1 = yn + (h/2)*(k1+k2)
    yn = yn1
    xn += h
    xn1 += h
  print("y(%.4f) = % .4f" % (xf, yn))

euleraper(xi,xf,yi,h)
```

```python
import numpy as np

def f(x,y):
  return x -y + 2

def Heun(y0,a,b,h):
  n = (b-a)/h
  n = int(n)
  x = np.arange(a,(b+h),h)
  y = np.zeros(n+1)
  y[0] = y0
  for i in range(0,n,1):
    k1 = f(x[i],y[i])
    k2 = f((x[i] + h), (y[i] + (k1*h)))
    k = (k1+k2)/2
    y[i+1] = y[i]+k*h
  return x,y

print(Heun(0,0,2.5,0.5))
```
