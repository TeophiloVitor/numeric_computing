**Plot da interpolação em pontos**

**Cálculo do centro de massa pelo método do simpson 3/8**

```python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('lab5att.jpg')

t = img.shape
fig, ax = plt.subplots(figsize=(10, 10))
def f(k):
  x = np.linspace(41, 1983, 10)
  y = [793,753,693,911,1077,1205,1259,1249,1169,645]
  y2 = [793,207,319,415,263,99,25,27,139,645]
  y1 = interpolacaoLagrange(k,x,y)
  y3 = interpolacaoLagrange(k , x , y2)
  deltay = y1 - y3

  return deltay 

#Simpson 3/8

a = 41
b = 1983
n = 9
h = (b - a)/n
soma = 0
somax = 0
somay = 0
if n%3 != 0:
  n = n - 1
for i in range(n + 1):
  if i == 0 or i == n:
    soma = soma + f(a)
    somax += f(a)*a
    somay += f(a)**2
  elif i%3 == 0:
    soma = soma + 2*f(a)
    somax += 2*f(a)*a
    somay += 2*(f(a)**2)
  else:
    soma = soma + 3*f(a)
    somax += 3*f(a)*a
    somay += 3*(f(a)**2)
  a = a + h
Area_Chinelo = (3*h/8)*soma
xcm_simpson2 = (3*h/8)*somax/Area_Chinelo
ycm_simpson2 = (3*h/8)*somay/Area_Chinelo
print("Area da figura em px² = " , Area_Chinelo)
print("Xcm por Simpson 3/8 = " , xcm_simpson2 )

ax.plot(xcm_simpson2 , Ycm, '.', linewidth=5, markersize=20, color='g')

ax.imshow(img, extent=[0, t[1], 0, t[0]])
```
